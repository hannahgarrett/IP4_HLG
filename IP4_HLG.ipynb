```python
# 5
#convert to string and use .replace
car_sales.head()
```

```python
car_sales.info()
```

```python
car_sales['Price'] = car_sales['Price'].str.replace(',', '')
car_sales['Price'] = car_sales['Price'].str.replace('$', '')
car_sales['Price'] = car_sales['Price'].astype(int)
car_sales.head()
```

```python
#6
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

```python
#7
periods=len(car_sales)
car_sales['Sale Date']=pd.date_range(start='2023-02-15',
                                     periods=periods,
                                    freq='D')
car_sales.head()
```

```python
#8
plt.plot(car_sales['Price'],
        car_sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs total sales line plot')
plt.show()
```

```python
#9
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price scatter plot')
plt.show()
```

```python
#10
plt.bar(car_sales['Make'],car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer bar plot')
plt.show()
```

```python
#11
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of car make')
plt.show()
```

```python
#12
fig=plt.figure(figsize=(9,7))

ax1=fig.add_subplot(221) 
ax1.set_title("Price vs total sales line plot")

ax1.plot(car_sales['Price'],
        car_sales['Total Sales'])

ax2=fig.add_subplot(222)
ax2.set_title('Odometer vs Price scatter plot')
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])


ax3=fig.add_subplot(223)
ax3.set_title('Make vs Odometer bar plot')
ax3.bar(car_sales['Make'],car_sales['Odometer (KM)'])


ax4=fig.add_subplot(224)
ax4.set_title('Histogram of make')
ax4.hist(car_sales['Make'])

plt.show()
```
